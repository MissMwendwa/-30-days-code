# Scatter 3D Plots

```python
#Imports
import seaborn as seaborn
import matplotlib.pyplot as plot
import numpy as np
```

```python
# Creating the data
mean = 3
number =50

x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)
```

```python
# Plotting the 3D scatter
plot.figure(figsize=(6,5))
axes = plot.axes(projection='3d')
```

```python
print(type(axes))
axes.scatter3D(x1, y1, z1)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
plot.show()
```

```python
seaborn.set_style("darkgrid")

mean = 3
number =50

x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

plot.figure(figsize=(6,5))
axes = plot.axes(projection='3d')
print(type(axes))
axes.scatter3D(x1, y1, z1)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
plot.show()
```

```python
# Increasing the size of the scatter plots
seaborn.set_style("darkgrid")

mean = 3
number =50

x1 = np.random.normal(mean, 1, size=number)
y1 = np.random.normal(mean, 1, size=number)
z1 = np.random.normal(mean, 1, size=number)

plot.figure(figsize=(6,5))
axes = plot.axes(projection='3d')
print(type(axes))
axes.scatter3D(x1, y1, z1,s=100)

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_zlabel('z')
plot.show()
```

```python
# A spiral Plot
import seaborn as sb
sb.set_style("whitegrid")

OMEGA = 2

Z1 = np.linspace(0, 10, 100)
X1 = np.cos(OMEGA*Z1)
Y1= np.sin(OMEGA*Z1)

plot.figure(figsize=(6,5))
axes = plot.axes(projection='3d')
axes.plot3D(X1,Y1,Z1)
# keeps padding between figure elements
plot.tight_layout()
plot.show()
```

```python
# a 3D plot of a map

def FUNC_Z(x, y):
    return 50 - (x**2 + y**2)


sb.set_style('whitegrid')
N = 50

X_VAL = np.linspace(-5, 5, N)
Y_VAL = np.linspace(-5, 5, N)

X1, Y1 = np.meshgrid(X_VAL, Y_VAL)

Z1 = FUNC_Z(X1, Y1)

axes = plot.axes(projection='3d')
axes.plot_surface(X1, Y1, Z1)
plot.show()
```
