# Seaborn HeatMaps


```python
# Imports
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
  
```

```python
# generating 2-D 10x10 matrix of random numbers
data = np.random.randint(low = 1,
                         high = 100,
                         size = (10, 10))

data
```

```python
# plotting the heatmap
hm = sn.heatmap(data = data)
plt.show()
```

```python
 
# setting the parameter values
vmin = 30
vmax = 70
  
# plotting the heatmap
hm = sn.heatmap(data=data,
                vmin=vmin,
                vmax=vmax)
  
# displaying the plotted heatmap
plt.show()
```

```python
cmap = "tab20"
  
# plotting the heatmap
hm = sn.heatmap(data=data,
                cmap=cmap)
  
# displaying the plotted heatmap
plt.show()
```

```python
cmap = "tab20"
center = 0
  
# plotting the heatmap
hm = sn.heatmap(data=data,
                cmap=cmap,
                center=center)
  
# displaying the plotted heatmap
plt.show()
```

```python
# setting the parameter values
annot = True
  
# plotting the heatmap
hm = sn.heatmap(data=data,
                annot=annot)
  
# displaying the plotted heatmap
plt.show()
```
