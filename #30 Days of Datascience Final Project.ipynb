```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset
data_one = pd.read_csv('adm_data.csv')
```

```python
data_one
```

```python
# first bar plot
plot = data_one['TOEFL Score']
plt.bar(range(len(plot)), plot, width=0.3, color='black')
plt.show()
```

```python
# Second a scatter plot
plot_2 = data_one['University Rating']
plt.scatter(plot, plot_2, color='black')
plt.show()
```

```python
# Line plot
plt.plot(plot, plot_2, color='purple')
plt.show()
```
