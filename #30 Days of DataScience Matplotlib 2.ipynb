# Scatter Plots and Line graphs

In this exercise, I will be doing two types of plots. The scatter plot and the Line plots. Let's do it


```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the data
data = np.array([40, 35, 39, 29, 10, 4, 16, 17])
data_2 = np.array([12, 14, 8, 20, 17, 18, 16, 12])
```

```python
# Line Graph
plt.plot(data, data_2, color='blue')
plt.show()
```

```python
# scatter plot
plt.scatter(data, data_2, color='green')
plt.show()
```

```python
# Plotting y=x**2 + 1
x = np.array([2, 4, 6, 8, 10, 12, 14, 16])
y = (x**2) + 1
```

```python
# Plotting
plt.plot(x, y, color = 'green')
plt.show()
```

```python
# Scatter
plt.scatter(x, y, color='orange')
plt.show()
```
