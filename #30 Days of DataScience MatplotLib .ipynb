```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
```

```python
# Plotting X vs Y
plt.plot(x, y, color="green")
plt.show()
```

```python
z = np.sin(x)
```

```python
# plotting x, y and z
plt.plot(x, y, color='Green')
plt.plot(x, z, color='Maroon')

plt.show()
```

```python
# Plotting sin
x = np.linspace(0, 10, 100)
v = np.tan(x)
```

```python
#plotting
plt.plot(x, y, color='orange')
plt.plot(x, z, color="red")
plt.plot(x, v, color='blue')

plt.show()
```

```python
# Plotting the graph y=x**2

new = np.array([12, 14, 16, 18, 20, 22, 24 ])
new_y = (new**2)
```
