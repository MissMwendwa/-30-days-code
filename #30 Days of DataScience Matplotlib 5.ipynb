# Bar Graphs from Functions


```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the data specifiers
num = 4
data = np.arange(num)
width = 0.25
```

```python
# Bar graphs
rent = [65000, 75000, 90000, 120000]
bar1 = plt.bar(data, rent, width, color = 'orange')

morgage = [45000, 65000, 78000, 140000]
bar2 = plt.bar(data, morgage, width, color = 'g')

no_of_apartments = [26000, 17000, 38000, 500]
bar3 = plt.bar(data, no_of_apartments, width, color = 'black')

value = [27000, 17000, 38000, 50000]
bar4 = plt.bar(data, no_of_apartments, width, color = 'blue')


plt.xlabel("House Hunting")
plt.ylabel('Market Range')
plt.title("Housing 2016")

plt.xticks(data+width, ['March 2016', 'June 2016', 'September 2016', 'November 2016'])
plt.legend( (bar1, bar2, bar3, bar4), ('Rent', 'Morgage', 'Apartments', 'Value') )
plt.show()
```

```python
# Second Bar Graph
num1 = 5

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(num1)
width = 0.35

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width, yerr = boyStd)
p2 = plt.bar(ind, girls, width,
			bottom = boys, yerr = girlStd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))

plt.show()
```
