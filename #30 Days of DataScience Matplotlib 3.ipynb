# MatplotLib Mathematical Equation Plots

In this particular notebook, My hope is to cover most of the equations I can recall

```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create the Data
x = np.array([2,4,6,8,10,12,14])
y = x*2
```

```python
# Plotting y=x*2 Algorithm
plt.plot(x, y, color='Purple')
plt.show()
```

```python
#Creating the data
x1 = np.array([1,2,4,5,7,2,7,6])
y1 = np.array([3,4,7,5,8,9,1,6])
```

```python
# Plotting y=x**2
plt.scatter(x1, y1, color='green')
plt.show()
```

```python
#Plotting 2 dimensional arrays
x2 = np.array([1,2,5,6,8,3,6,2])
y2 = np.array([4,6,7,2,9,4,6,7])
z = np.array([3,2,5,8,3,7,5,2])
```

```python
# Scatter plots
plt.scatter(x2, y2, color='orange')
plt.scatter(x2, z, color='blue')
plt.show()
```

```python
# First Bar Graph
data = [5, 35, 25, 85, 60, 55]
plt.bar(range(len(data)), data, color = 'orange')
plt.show()
```
